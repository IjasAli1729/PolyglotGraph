```python
from langgraph.graph import StateGraph, END,START
from typing import TypedDict
import os
from langchain_google_genai  import ChatGoogleGenerativeAI
from dotenv import load_dotenv, find_dotenv
load_dotenv(find_dotenv())
model = ChatGoogleGenerativeAI(
    model= "gemini-2.5-flash"
)
```

```python
class state (TypedDict):
    text: str
    french: str
    spanish: str
    japanese: str
    chinese: str
    combined_text = str
```

```python
def french_translate(state:state) -> dict:
    response = model.invoke(f"translate the given text to franch: \n\n{state['text']}")
    return{"french": response.content.strip()}


def spanish_translate(state:state) -> dict:
    response = model.invoke(f"translate the given text to spanish: \n\n{state['text']}")
    return{"spanish": response.content.strip()}

def japanese_translate(state:state) -> dict:
    response = model.invoke(f"translate the given text to japanese: \n\n{state['text']}")
    return{"japanese": response.content.strip()}

def chinese_translate(state:state) -> dict:
    response = model.invoke(f"translate the given text to chinese: \n\n{state['text']}")
    return{"chinese": response.content.strip()}


def aggregator(state:state) -> dict:
    combined = f"Orginal Text :{state['text']}\n\n"
    combined += f"French : {state['french']}\n\n"
    combined += f"Spanish : {state['spanish']}\n\n"
    combined += f"Chinese : {state['chinese']}\n"
    return{"combined_text":combined}
```

```python
graph = StateGraph(state)
```

```python
graph.add_node("French_translate",french_translate)
graph.add_node("Spanish_translate",spanish_translate)
graph.add_node("Japanese_translate", japanese_translate)
graph.add_node("Chinese_translate", chinese_translate)
graph.add_node("Aggregator", aggregator)
```

```python
graph.add_edge(START, "French_translate")
graph.add_edge(START,"Spanish_translate")
graph.add_edge(START,"Japanese_translate")
graph.add_edge(START,"Chinese_translate")
```

```python
graph.add_edge("French_translate","Aggregator")
graph.add_edge("Spanish_translate","Aggregator")
graph.add_edge("Japanese_translate","Aggregator")
graph.add_edge("Chinese_translate","Aggregator")
```

```python
graph.set_finish_point("Aggregator")
```

```python
app = graph.compile()
```

```python
input_text = {
    "text" : "hello, how was you day?"
}

result = app.invoke(input_text)
```
